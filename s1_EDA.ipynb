```python
import pandas as pd
```

```python
df=pd.read_csv("drugbank_clean.csv")
```

```python
null_stats = pd.DataFrame({
    'column': df.columns,
    'null_count': df.isnull().sum(),
    'null_percentage': df.isnull().mean() * 100
})


null_stats_sorted = null_stats.sort_values(by='null_percentage', ascending=True)
top_20_least_nulls = null_stats_sorted.head(20)


print(top_20_least_nulls)
top_20_columns = top_20_least_nulls['column'].tolist()
df_top20 = df[top_20_columns]
```

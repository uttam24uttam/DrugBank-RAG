```python
import pandas as pd
```

```python
df=pd.read_csv("drugbank_clean.csv")
df.head()
```

```python
# Count nulls
null_count = df.isnull().sum()

# Percentage of nulls
null_percentage = (null_count / len(df)) * 100

# Combine into a nice dataframe
null_summary = pd.DataFrame({
    "column": df.columns,
    "null_count": null_count.values,
    "null_percentage": null_percentage.values
})

# Sort by most missing
null_summary = null_summary.sort_values(by="null_percentage", ascending=False)

# Display
print(null_summary)
```

```python
columns_to_keep = [
    # Identifiers & basic info
    "drugbank-id", "name", "cas-number", "unii",
    
    # Pharmacological properties (key ones)
    "indication", "mechanism-of-action", "pharmacodynamics", "toxicity",
    "absorption", "metabolism", "half-life", "clearance",
    
    # Codes & labels
    "atc-codes", "fda-label", "msds", "food-interactions", "drug-interactions",
    
    # Molecular / structural info
    "targets", "enzymes", "carriers", "transporters"
]
# Keep only the columns that exist in the CSV (some might be missing)
columns_to_keep_existing = [col for col in columns_to_keep if col in df.columns]

df_selected = df[columns_to_keep_existing]
```

```python

# Assuming df_selected contains only your chosen columns
# Count nulls
null_count = df_selected.isnull().sum()

# Percentage of nulls
null_percentage = (null_count / len(df_selected)) * 100

# Combine into a dataframe
null_summary = pd.DataFrame({
    "column": df_selected.columns,
    "null_count": null_count.values,
    "null_percentage": null_percentage.values
})

# Sort by most missing
null_summary = null_summary.sort_values(by="null_percentage", ascending=False)

# Display
print(null_summary)
```

```python
null_stats = pd.DataFrame({
    'column': df.columns,
    'null_count': df.isnull().sum(),
    'null_percentage': df.isnull().mean() * 100
})

# Sort by null percentage ascending (least nulls first)
null_stats_sorted = null_stats.sort_values(by='null_percentage', ascending=True)

# Get top 20 columns with least nulls
top_20_least_nulls = null_stats_sorted.head(20)

# Print
print(top_20_least_nulls)

# Optionally, get the column names
top_20_columns = top_20_least_nulls['column'].tolist()

# Filter the original df to only these columns
df_top20 = df[top_20_columns]
```

```python
# Count non-null columns per row
df_selected['non_null_count'] = df_selected.notnull().sum(axis=1)

# Sort by non-null count descending (most complete drugs first)
df_sorted = df_selected.sort_values(by='non_null_count', ascending=False)

# Pick top 10k drugs
top_10k_drugs = df_sorted.head(10000).drop(columns=['non_null_count'])
```

```python
top_10k_drugs
```

```python
# Count of nulls per column
null_count = top_10k_drugs.isnull().sum()

# Percentage of nulls per column
null_percentage = (null_count / len(top_10k_drugs)) * 100

# Combine into a single DataFrame for clarity
null_summary = pd.DataFrame({
    "column": top_10k_drugs.columns,
    "null_count": null_count,
    "null_percentage": null_percentage
}).sort_values(by="null_percentage")

print(null_summary)
```

```python
import pandas as pd

# Count nulls for all columns
null_counts = df.isnull().sum().sort_values()
null_percentage = (null_counts / len(df)) * 100

# Combine in a DataFrame for easy viewing
null_summary = pd.DataFrame({
    'column': null_counts.index,
    'null_count': null_counts.values,
    'null_percentage': null_percentage.values
})

# Pick top 25 columns with least nulls (adjust 20–30 as needed)
top_columns = null_summary['column'].iloc[:25].tolist()
print("Top 25 columns with least nulls:", top_columns)

# Filter df to keep only these columns
df_top_cols = df[top_columns]
```

```python
df_top_cols
```

```python
# Column-wise null % 
col_null_percentage = (df_top_cols.isnull().sum() / len(df_top_cols)) * 100

col_null_df = pd.DataFrame({
    'column': df_top_cols.columns,
    'null_percentage': col_null_percentage.values
}).sort_values('null_percentage')

print("Column-wise null %:")
print(col_null_df)
```

```python
# Row-wise null % 
row_null_percentage = (df_top_cols.isnull().sum(axis=1) / len(df_top_cols.columns)) * 100

row_null_df = pd.DataFrame({
    'drug_index': df_top_cols.index,
    'null_percentage': row_null_percentage.values
}).sort_values('null_percentage')

print("Row-wise null % (first 10 rows):")
print(row_null_df.head(10000))
```

```python
# Remove only the 'toxicity' column
df_final = df_top_cols.drop(columns=['toxicity'])

# Check resulting columns
print(df_final.columns)
```
